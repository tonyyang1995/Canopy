```python
print("step1: define args and load dataset")
excel_file_path = './Data_Collection_V3.xlsx'
sheet_page = "Sheet1" # sheet page name
label = "Strength" # the label for the pred column
ridge_save_path = "results_test_nano_other/ridge_regression_"
rf_save_path = "result_test_nano_other/rf_regression"
svm_save_path = "results_test_nano_other/svm_regression"
xgb_save_path = "results_test_nano_other/xgb_regression"
ann_save_path = "results_test_nano_other/ann_regression"

box_plot_colors = sns.color_palette("Set2")
kfold = 10
split_rate=0.1
facecolor = "#FF8C94"
patchcolor = 'red'
alpha = 0.3
# xlim = (0, 120)
# ylim = (0, 120)
xlim = None
ylim = None

df = read_excel(excel_file_path, sheet_page)
df = df.drop('Data', axis=1)
column_names = list(df.columns)
print(column_names)
save_paths = []
for feature in column_names[1:13]:
    sp = draw_hex_contour_plot(df, feature, label, "#4CB391")
    save_paths.append(sp)
```

```python
excel_file_path = './Concrete_Data_Test.xlsx'
test_df = read_excel(excel_file_path, sheet_page)
lens = test_df.shape[0]
test_df["SF"] = [0] * lens
test_df["nano-TiO2"] = [0] * lens
test_df['nano-SiO2'] = [0] * lens
test_df["TempdegC"] = [0] * lens
test_df_select = test_df[list(df.columns)]
print(test_df.columns)
```

```python
test_columns = list(test_df_select.columns)
x_test = test_df[test_columns[:-1]]
y_test = test_df[["Strength"]]
```

```python
# 2. draw joint plots
print("step2: analysis dataset and draw plots")
row = 4
col= 3
count = 0
fig, axs = plt.subplots(row,col, figsize=(20,20))
for i in range(row):
    for j in range(col):
        show_joint_plots(save_paths[count], axs[i][j])
        count += 1
plt.tight_layout()
# plt.show()
```

```python
def draw_heatmap2(df, drop_column=None, cmap='coolwarm'):
    new_df = df.fillna(0)
    if drop_column is not None:
        new_df.drop(columns=drop_column, inplace=True)
    corr_matrix = new_df.corr()
    sns.heatmap(corr_matrix, annot=True, cmap=cmap)
# 3. draw heatmap
print("step3: drawing heatmap")
plt.close("all")
plt.figure(figsize=(15, 15))
# cmap options: ["YlGnBu", "Blues", "coolwarm", "BuPu", "Greens", "Oranges", "Reds", "Purples", "YlOrBr"]
draw_heatmap2(df, cmap='Greens')
plt.title("Heatmap")
if not os.path.exists("results_test_nano_other"):
    os.makedirs("results_test_nano_other")
plt.savefig('results_test_nano_other/heatmap.png')
# plt.show()
```

```python
column_names = df.columns
x_train = df[column_names[:-1]]
y_train = df[[column_names[-1]]]
```

```python
from sklearn import linear_model
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
import xgboost as xgb
```

```python
import importlib
import utils
importlib.reload(utils.model)
from utils.model import calculate_rmse, calculate_rmae
from utils.model import r2_score, calculate_RSR
from utils.model import calculate_mape, calcuate_nmbe
```

```python
# ridge
model = linear_model.RidgeCV(alphas=np.logspace(-6, 6, 13))
model.fit(x_train, y_train)
pred = model.predict(x_test)
rmse = calculate_rmse(y_test, pred)
mae = calculate_rmae(y_test, pred)
r2 = r2_score(y_test, pred)
rsr = calculate_RSR(y_test, pred)
mape = calculate_mape(y_test, pred)
nmbe = calcuate_nmbe(y_test, pred)
print("ridge regression")
print(f"rmse: {rmse}")
print(f"mae: {mae}")
print(f"r2: {r2}")
print(f"r2r: {rsr}")
print(f"mape: {mape}")
print(f"nmbe: {nmbe}")
```

```python
# rf
model = RandomForestRegressor(random_state=42)
model.fit(x_train, y_train)
pred = model.predict(x_test)
rmse = calculate_rmse(y_test, pred)
mae = calculate_rmae(y_test, pred)
r2 = r2_score(y_test, pred)
rsr = calculate_RSR(y_test, pred)
mape = calculate_mape(y_test, pred)
nmbe = calcuate_nmbe(y_test, pred)
print("random forest")
print(f"rmse: {rmse}")
print(f"mae: {mae}")
print(f"r2: {r2}")
print(f"r2r: {rsr}")
print(f"mape: {mape}")
print(f"nmbe: {nmbe}")
```

```python
model = xgb.XGBRegressor(n_estimators=1000)
model.fit(x_train, y_train)
pred = model.predict(x_test)
rmse = calculate_rmse(y_test, pred)
mae = calculate_rmae(y_test, pred)
r2 = r2_score(y_test, pred)
rsr = calculate_RSR(y_test, pred)
mape = calculate_mape(y_test, pred)
nmbe = calcuate_nmbe(y_test, pred)
print("xgboost")
print(f"rmse: {rmse}")
print(f"mae: {mae}")
print(f"r2: {r2}")
print(f"r2r: {rsr}")
print(f"mape: {mape}")
print(f"nmbe: {nmbe}")
```

```python
model = MLPRegressor(hidden_layer_sizes=(32, 128, 128), max_iter=1000, random_state=42, tol=1e-6, batch_size=64)
model.fit(x_train, y_train)
pred = model.predict(x_test)
rmse = calculate_rmse(y_test, pred)
mae = calculate_rmae(y_test, pred)
r2 = r2_score(y_test, pred)
rsr = calculate_RSR(y_test, pred)
mape = calculate_mape(y_test, pred)
nmbe = calcuate_nmbe(y_test, pred)
print("ANN")
print(f"rmse: {rmse}")
print(f"mae: {mae}")
print(f"r2: {r2}")
print(f"r2r: {rsr}")
print(f"mape: {mape}")
print(f"nmbe: {nmbe}")
```
